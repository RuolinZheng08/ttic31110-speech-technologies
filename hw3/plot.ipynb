```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.stats as ss
```

```python
ps = np.arange(0.1, 0.6, 0.1)
for p in ps:
    x = np.arange(ss.geom.ppf(0.01, p), ss.geom.ppf(0.99, p))
    plt.plot(ss.geom.pmf(x, p), label='p={:.2f}'.format(p))
plt.legend()
plt.title('Geometric')
plt.show()
```

```python
n = 2
ps = np.arange(0.1, 0.6, 0.1)
for p in ps:
    x = np.arange(ss.nbinom.ppf(0.01, n, p), ss.nbinom.ppf(0.99, n, p))
    plt.plot(ss.nbinom.pmf(x, n, p), label='p={:.2f}, n={}'.format(p, n))
plt.legend()
plt.title('Negative Binomial')
plt.show()
```

```python
ps = [0.3, 0.6]
n = 2
for p in ps:
    x = np.arange(ss.geom.ppf(0.01, p), ss.geom.ppf(0.99, p))
    plt.plot(ss.geom.pmf(x, p), label='p={:.2f}'.format(p))
    x = np.arange(ss.nbinom.ppf(0.01, n, p), ss.nbinom.ppf(0.99, n, p))
    plt.plot(ss.nbinom.pmf(x, n, p), label='p={:.2f}, n={}'.format(p, n))
plt.legend()
plt.title('Geometric & Neg Binomial Comparison')
plt.show()
```

```python
p = 0.5
for n in range(1, 8):
    x = np.arange(ss.nbinom.ppf(0.01, n, p), ss.nbinom.ppf(0.99, n, p))
    plt.plot(ss.nbinom.pmf(x, n, p), label='p={:.2f}, n={}'.format(p, n))
plt.legend()
plt.title('Negative Binomial')
plt.show()
```

```python
n_samples = 1000
for prob in [0.3, 0.5, 0.7, 0.9]:
    res = prob * np.random.geometric(0.6, n_samples) + \
    (1 - prob) * np.random.geometric(0.2, n_samples)
    plt.hist(res, alpha=0.75, label='a={}'.format(prob), density=True)
plt.legend()
plt.title('Two tied states, p1=0.6, p2=0.2')
plt.show()
```

```python
for prob in [0.3, 0.5]:
    for prob2 in [0.1, 0.3]:
        res = prob * np.random.geometric(0.6, n_samples) + \
        prob2 * np.random.geometric(0.2, n_samples)
        (1 - prob - prob2) * np.random.geometric(0.5, n_samples)
        plt.hist(res, alpha=0.75, label='a1={},a2={}'.format(prob, prob2), density=True)
plt.legend()
plt.title('Three tied states, p1=0.6, p2=0.2, p3=0.5')
plt.show()
```
